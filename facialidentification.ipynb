# No of celebs are -> 100

```python
#data transformation
for i in range(len(x)):
    x[i]=cv2.resize(x[i],(224, 224)).astype(np.float32)

#train test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.25)
del(x)
del(y)
```

```python
len(X_train)
```

```python
del(celebNames)
```

```python
#crreating model

model =Sequential()
#model.add(VGG16(include_top=False,weights='imagenet'))
model.add(VGG16(include_top=False,weights='imagenet',input_shape=(224, 224, 3)))
model.add(Conv2D(filters=100,kernel_size=(7,7), activation="relu"))
model.add(Flatten())
model.add(Dense(100, activation='softmax'))
model.compile( optimizer='adam', loss="sparse_categorical_crossentropy", metrics=['accuracy'] )
```

```python
#model=VGG16(include_top=False)
model.summary()
```

```python
for i in range(0,7521-500,500):
    print("###################################################################################")
    print()
    print(i)
    print()
    print("###################################################################################")
    model.fit(x=np.array(X_train[i:i+500]),y=np.array(y_train[i:i+500]),epochs=5,batch_size=25)
    model.save('model.h5')
```
